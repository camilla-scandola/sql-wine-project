```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import os

# Fetch the dataset
wine = fetch_ucirepo(id=109)

# Extract features (X) and target (y)
X = wine.data.features
y = wine.data.targets

# Combine them into one DataFrame
wine_df = pd.concat([X, y], axis=1)

# Define your save path
save_path = "ucimlrepo/wine_dataset.csv"

# (optional) Make sure the folder exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save to CSV
wine_df.to_csv(save_path, index=False)

print(f"Dataset saved successfully to:\n{save_path}")
```

```python
import pandas as pd
import unicodedata

# Input and output file paths
input_path = "https://github.com/camilla-scandola/sql-wine-project/blob/main/winemag-data-130k-v2.csv"
output_path = "https://github.com/camilla-scandola/sql-wine-study/winemag-data-clean.csv"

# Load the dataset
df = pd.read_csv(input_path, encoding="utf-8", low_memory=False)

# Remove accents from all text columns
def strip_accents(text):
    if isinstance(text, str):
        text = unicodedata.normalize('NFKD', text)
        return ''.join([c for c in text if not unicodedata.combining(c)])
    return text

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(strip_accents)

# Save the clean UTF-8 file
df.to_csv(output_path, index=False, encoding="utf-8")

print("Clean file saved at:", output_path)
```

```python
import os
import mysql.connector
from dotenv import load_dotenv

# load environment variables from .env during local development
load_dotenv()

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password=os.getenv("DB_PASSWORD", "password"),  # default fallback for dev
    database="sql_wine_study"
)
```

```python
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap

# query for average chemical values
uci_query = """
SELECT
    class,
    ROUND(AVG(Alcohol), 2) AS avg_alcohol,
    ROUND(AVG(Malicacid), 2) AS avg_malic_acid,
    ROUND(AVG(Color_intensity), 2) AS avg_color_intensity,
    ROUND(AVG(Total_phenols), 2) AS avg_phenols,
    ROUND(AVG(Flavanoids), 2) AS avg_flavanoids,
    ROUND(AVG(Proline), 2) AS avg_proline
FROM wine_dataset
GROUP BY class
ORDER BY class;
"""

uci_df = pd.read_sql(uci_query, conn)

# style DataFrame

# define 3-color gradient
colors = ["#EED3DC", "#F3C1D4", "#D3A1B3"]  # light → medium → dark pink-red
three_color_cmap = LinearSegmentedColormap.from_list("three_color_map", colors)

# apply gradient per column
styled_uci = (
    uci_df.style
    .background_gradient(cmap=three_color_cmap, axis=0, subset=uci_df.columns[1:])
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
)

styled_uci

```

```python
# query for level map
wine_class_query = """
SELECT 
    class,
    alcohol_quantity,
    malic_acid_level,
    phenols_level,
    flavanoids_level,
    color_intensity_level
FROM wine_class_characteristics
ORDER BY class;
"""

wine_class_df = pd.read_sql(wine_class_query, conn)

# style DataFrame

# define a color map for categorical levels
color_map = {
    "Low": "#EED3DC",      # light pink
    "Medium": "#F3C1D4",   # medium rose
    "High": "#D3A1B3"      # deep red
}

# function to apply color based on value
def color_levels(val):
    color = color_map.get(val, "white")
    return f"background-color: {color}; text-align: center;"

# apply the styling
styled_wine_class = (
    wine_class_df.style
    .applymap(color_levels, subset=wine_class_df.columns[1:])
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
)

styled_wine_class
```

```python
import matplotlib.pyplot as plt
import numpy as np

# query for average chemical values
query = """
SELECT 
    class,
    ROUND(AVG(Alcohol), 2) AS avg_alcohol,
    ROUND(AVG(Malicacid), 2) AS avg_malic_acid,
    ROUND(AVG(Color_intensity), 2) AS avg_color_intensity,
    ROUND(AVG(Total_phenols), 2) AS avg_phenols,
    ROUND(AVG(Flavanoids), 2) AS avg_flavanoids
FROM wine_dataset
GROUP BY class;
"""
df = pd.read_sql(query, conn)

# prepare data
classes = df['class'].astype(str)
values = df.drop('class', axis=1)
colors = ['#A45C6B', '#B76E79', '#C98C8C', '#D8A7A7', '#EBC4C4']  # warm wine tones

# plot (grouped bar chart)
x = np.arange(len(classes))
width = 0.15

plt.figure(figsize=(10, 6))
for i, col in enumerate(values.columns):
    plt.bar(x + (i - 2) * width, values[col], width, label=col.replace('_', ' '), color=colors[i])

# 5. Minimal styling
plt.xlabel('Wine Class', fontsize=11)
plt.ylabel('Average Value', fontsize=11)
plt.title('Average Chemical Composition by Wine Class', fontsize=13, fontweight='bold')
plt.xticks(x, classes)
plt.legend(title='Chemical Component', frameon=False)
plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# query to get average rating and price by variety
query = """
SELECT 
    variety,
    province,
    ROUND(AVG(points), 1) AS avg_rating,
    ROUND(AVG(CAST(price AS DECIMAL(10,2))), 2) AS avg_price
FROM winemag_data_clean
WHERE country = 'Italy'
GROUP BY variety, province
ORDER BY avg_rating DESC
LIMIT 10;
"""

df = pd.read_sql(query, conn)

# prepare data
x = np.arange(len(df))
width = 0.35

# color palette
colors = ['#A45C6B', '#D8A7A7']

# create the bar chart
plt.figure(figsize=(10,6))
plt.bar(x - width/2, df['avg_rating'], width, label='Average Rating', color=colors[0])
plt.bar(x + width/2, df['avg_price'], width, label='Average Price', color=colors[1])

# style and labels
plt.xlabel('Wine Variety', fontsize=11)
plt.ylabel('Value', fontsize=11)
plt.title('Average Rating and Price by Italian Wine Variety', fontsize=13, fontweight='bold')
plt.xticks(x, df['variety'], rotation=45, ha='right')
plt.legend(frameon=False)
plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# query for the 3 Piemonte wines
query = """
SELECT 
    variety,
    province,
    ROUND(AVG(points), 1) AS avg_rating,
    ROUND(AVG(CAST(price AS DECIMAL(10,2))), 2) AS avg_price
FROM winemag_data_clean
WHERE variety IN ('Nebbiolo', 'Dolcetto', 'Barbera')
GROUP BY variety, province
ORDER BY avg_rating DESC;
"""

df = pd.read_sql(query, conn)

# plot — dual-line style
plt.figure(figsize=(8,6))

# line colors (burgundy palette)
rating_color = '#A45C6B'
price_color = '#D8A7A7'

plt.plot(df['variety'], df['avg_rating'], marker='o', color=rating_color, linewidth=2.5, label='Average Rating')
plt.plot(df['variety'], df['avg_price'], marker='o', color=price_color, linewidth=2.5, label='Average Price')

# style and formatting
plt.title('Average Rating and Price - Nebbiolo, Dolcetto & Barbera', fontsize=13, fontweight='bold')
plt.xlabel('Wine Variety', fontsize=11)
plt.ylabel('Value', fontsize=11)
plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```
